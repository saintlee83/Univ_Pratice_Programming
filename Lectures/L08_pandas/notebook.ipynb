```python
import pandas as pd

# 1) 기본 예제 데이터
df = pd.DataFrame({
    "class": ["A", "A", "A", "B", "B", "C", "C", "C", "C"],
    "sex":   ["F", "M", "F", "M", "F", "M", "F", "F", "M"],
    "score": [80, 70, 90, 60, 75, 88, 92, 85, 73],
    "age":   [20, 22, 21, 24, 23, 26, 22, 21, 27],
})

df
```

```python
temp = df.groupby("class")

print(type(temp))
print(temp)

# temp.shape # DataFrameGroupBy

# AttributeError: 'DataFrameGroupBy' object has no attribute 'shape'

temp.mean(numeric_only=True)

# temp.mean()
# TypeError: agg function failed [how->mean,dtype->object]
```

```python
temp = df.groupby("class").mean(numeric_only=True)

print(type(temp))
print(temp)
```

```python
val = df.groupby("class").mean(numeric_only=True)

print(val.__repr__)
```

```python
temp = df.groupby("class")["score"].mean()

print(type(temp))
print(temp)
```

```python
temp = df.groupby("class")[["score"]].mean()

print(type(temp))
print(temp)
```

```python
# 다중 컬럼명
temp = df.groupby(["class","sex"])["score"].mean()

print(type(temp))
print(temp)

temp.loc["A"]
```

```python
row_groups = [i % 2 for i in range(len(df))]
print(df.groupby(row_groups)["score"].mean())
```
