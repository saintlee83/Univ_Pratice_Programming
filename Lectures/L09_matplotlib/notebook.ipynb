```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)
df = iris.frame
df["species"] = df["target"].map(dict(enumerate(iris.target_names)))

# 첫 20개 샘플만
df.iloc[:20, :4].plot.line(
    figsize=(7, 4),
    title="Line Plot of First 20 Samples",
    grid=True,
    style="--o"
)
plt.show()
```

```python
df.head()
```

```python
df.tail()
```

```python
df.groupby("species").mean().plot.bar(
    figsize=(7, 5),
    stacked=False,
    title="Mean Feature Values by Species",
    rot=0
)  # 목적: 그룹별 평균 값 비교
plt.show()
```

```python
df.groupby("species").mean().plot.bar(
    figsize=(7, 5),
    stacked=True,
    title="Mean Feature Values by Species",
    rot=0
)  # 목적: 그룹별 평균 값 비교
plt.show()
```

```python
df.groupby("species").mean().plot.barh(
    figsize=(7, 5),
    stacked=True,
    title="Mean Feature Values by Species",
    rot=0
)  # 목적: 그룹별 평균True 값 비교
plt.show()
```

```python
df.plot.hist(
    y="petal length (cm)",
    bins=20,
    alpha=0.7,
    grid=True,
    title="Histogram of Petal Length"
)  # 목적: 특정 feature의 분포 확인
plt.show()
```

```python
df.iloc[:, :4].plot.kde(
    figsize=(7, 5),
    title="KDE of Iris Features",
    lw=2  # line width
)
```

```python
df.iloc[:, :4].plot.box(
    figsize=(6, 5),
    grid=True,
    title="Box Plot of Iris Features",
    rot=50,
    sym="o"
)  # 목적: 분포 범위, 이상치 탐지
plt.show()
```

```python
df.iloc[:20, :4].plot.area(
    alpha=0.4,
    stacked=False,
    figsize=(7, 4),
    title="Area Plot (Non-Stacked)"
)
plt.show()  # 목적: feature 크기 비교
```

```python
df.iloc[:20, :4].plot.line(
    alpha=0.4,
    stacked=False,
    figsize=(7, 4),
    title="Area Plot (Non-Stacked)"
)
plt.show()  # 목적: feature 크기 비교
```

```python
df.plot.scatter(
    x="sepal length (cm)",
    y="petal length (cm)",
    c="petal width (cm)",
    colormap="viridis",
    alpha=0.7,
    title="Scatter Plot: Sepal vs Petal"
) # 목적: feature 간 상관관계 확인
plt.show()
```

```python
ax = df.groupby("species")["petal length (cm)"].count().plot.pie(
    autopct="%.2f%%",
    ylabel="",
    startangle=90,
    title="Sample Count by Species"
) # 목적: 범주별 비율 비교
plt.show()
```
